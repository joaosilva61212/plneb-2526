Create a function that:
1. given a string “s”, reverse it


```python
s = "Hello World"
def inverter_string(s):
    return s[::-1]

inverter_string(s)


```

2. given a string “s”, returns how many “a” and “A” characters are present in it.

```python
def count_a(s):
    s =s.lower().count("a")
    return s

count_a(s)
```

3. given a string “s”, returns the number of vowels there are present in it.

```python
def count_vowels(s):
    count = 0
    vogais = "aeiou"
    s = s.lower()
    for char in s:
        if char in vogais:
            count += 1
    return count

count_vowels(s)                           
```

4. given a string “s”, convert it into lowercase.

```python
def minusculas(s):
    return s.lower()

minusculas(s)
```

5. given a string “s”, convert it into uppercase.

```python
def maiusculas(s):
    return s.upper()

maiusculas(s)
```

6. Verifica se uma string é capicua

```python
def palindromo(s):
    s = s.lower().replace(" ", "") #remover espaços
    return s == s[::-1]

palindromo(s)
```

7. Verifica se duas strings estão balanceadas (Duas strings, s1 e s2, estão
balanceadas se todos os caracteres de s1 estão presentes em s2.)

```python
s1 = "aeiou"
s2= "aeioufg"
def balanceamento(s1, s2):
    for char in s1:
        if char not in s2:
            return False
    return True

balanceamento(s1,s2)
```

8. Calcula o número de ocorrências de s1 em s2

```python
s1="ana"
s2="banana"
def count_ocorrencias(s1,s2):
    return s2.count(s1)

count_ocorrencias(s1,s2)
```
